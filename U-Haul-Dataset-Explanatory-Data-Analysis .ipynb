```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('/kaggle/input/u-haul-dataset-rental-rates/uhauloutput.csv')

# Removing the dollar sign and converting Rate to float for analysis
df['Rate'] = df['Rate'].replace('[\$,]', '', regex=True).astype(float)
```

## Summary statistics of the rates

```python
rate_summary = df['Rate'].describe()
print("Summary Statistics of Rental Rates")
print(rate_summary)
```

##  Pickup Location and calculating mean rate

```python
mean_rate_by_pickup = df.groupby('Pickup Location')['Rate'].mean().sort_values(ascending=False)
print("\nAverage Rental Rates by Pickup Location")
print(mean_rate_by_pickup.head())
```

##  Dropoff Location and calculating mean rate

```python
mean_rate_by_dropoff = df.groupby('Dropoff Location')['Rate'].mean().sort_values(ascending=False)
print("\nAverage Rental Rates by Dropoff Location")
print(mean_rate_by_dropoff.head())
```

## Truck Type and calculating mean rate

```python
# Grouping by Truck Type and calculating mean rate
mean_rate_by_truck_type = df.groupby('Truck Type')['Rate'].mean().sort_values(ascending=False)
print("\nAverage Rental Rates by Truck Type")
print(mean_rate_by_truck_type.head())
```

## Moving Type and calculating mean rate

```python
# Grouping by Moving Type and calculating mean rate
mean_rate_by_moving_type = df.groupby('Moving Type')['Rate'].mean().sort_values(ascending=False)
print("\nAverage Rental Rates by Moving Type")
print(mean_rate_by_moving_type.head())
```

## Average rates by pickup location

```python
# Plotting the average rates by pickup location
plt.figure(figsize=(12, 8))
sns.barplot(x=mean_rate_by_pickup.index, y=mean_rate_by_pickup.values)
plt.xticks(rotation=90)
plt.title('Average Rental Rates by Pickup Location')
plt.ylabel('Average Rate (USD)')
plt.xlabel('Pickup Location')
plt.show()
```

## Average rates by dropoff location

```python
# Plotting the average rates by dropoff location
plt.figure(figsize=(12, 8))
sns.barplot(x=mean_rate_by_dropoff.index, y=mean_rate_by_dropoff.values)
plt.xticks(rotation=90)
plt.title('Average Rental Rates by Dropoff Location')
plt.ylabel('Average Rate (USD)')
plt.xlabel('Dropoff Location')
plt.show()
```

##  Average rates by truck type

```python
# Plotting the average rates by truck type
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_rate_by_truck_type.index, y=mean_rate_by_truck_type.values)
plt.title('Average Rental Rates by Truck Type')
plt.ylabel('Average Rate (USD)')
plt.xlabel('Truck Type')
plt.show()
```

 ## Average rates by moving type

```python
# Plotting the average rates by moving type
plt.figure(figsize=(10, 6))
sns.barplot(x=mean_rate_by_moving_type.index, y=mean_rate_by_moving_type.values)
plt.title('Average Rental Rates by Moving Type')
plt.ylabel('Average Rate (USD)')
plt.xlabel('Moving Type')
plt.show()
```
